### Problem 1:
#### Generate data of size 250, 300 and 500 from bivariate and trivariate normal and standard normal distributions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Set the random seed for reproducibility
np.random.seed(42)

# Parameters for bivariate normal distribution
mean_bivariate = [0, 0]
cov_bivariate = [[1, 0.5], [0.5, 1]]  # Covariance matrix

# Parameters for trivariate normal distribution
mean_trivariate = [0, 0, 0]
cov_trivariate = [[1, 0.5, 0.3], [0.5, 1, 0.4], [0.3, 0.4, 1]]  # Covariance matrix

# Generate data
data_bivariate_250 = np.random.multivariate_normal(mean_bivariate, cov_bivariate, 250)
data_bivariate_300 = np.random.multivariate_normal(mean_bivariate, cov_bivariate, 300)
data_bivariate_500 = np.random.multivariate_normal(mean_bivariate, cov_bivariate, 500)

data_trivariate_250 = np.random.multivariate_normal(mean_trivariate, cov_trivariate, 250)
data_trivariate_300 = np.random.multivariate_normal(mean_trivariate, cov_trivariate, 300)
data_trivariate_500 = np.random.multivariate_normal(mean_trivariate, cov_trivariate, 500)

# Generate data from standard normal distribution
data_standard_normal_250 = np.random.randn(250)
data_standard_normal_300 = np.random.randn(300)
data_standard_normal_500 = np.random.randn(500)
```

```python
# Output the shapes to verify the sizes
print("Bivariate (250 samples): \n", data_bivariate_250[:20])
print("----------------------------------------")
print("Bivariate (300 samples): \n", data_bivariate_300[:20])
print("----------------------------------------")
print("Bivariate (500 samples): \n", data_bivariate_500[:20])
```

```python
print("Trivariate (250 samples): \n", data_trivariate_250[:20])
print("----------------------------------------")
print("Trivariate (300 samples): \n", data_trivariate_300[:20])
print("----------------------------------------")
print("Trivariate (500 samples): \n", data_trivariate_500[:20])
```

```python
print("Standard Normal (250 samples): \n", data_standard_normal_250[:20])
print("----------------------------------------")
print("Standard Normal (300 samples): \n", data_standard_normal_300[:20])
print("----------------------------------------")
print("Standard Normal (500 samples): \n", data_standard_normal_500[:20])
```

### Problem 2: 
#### Obtain the Maximum Likelihood Estimate of mean vector and variance covariance matrix for Weight, Hieght and MRI from the following data. Assume that the data comes from normal distribution. Also draw Q-Q Plot for these considered variables. 

```python
# Data
data = pd.read_excel("female.xlsx")

data
```

```python
# Calculate the Mean Vector
mean_vector = data.mean()
print("Mean Vector:")
print(mean_vector)
```

```python
# Calculate the Variance-Covariance Matrix
cov_matrix = data.cov()
print("\nVariance-Covariance Matrix:")
print(cov_matrix)
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

for col in data.select_dtypes(include=[float, int]).columns:  # Filter numeric columns
    plt.figure()
    if data[col].dropna().empty:  # Skip columns with only NaN values
        print(f"Skipping {col} as it contains only NaN values")
        continue
    stats.probplot(data[col].dropna(), dist="norm", plot=plt)  # Drop NaN values before plotting
    plt.title(f"Q-Q Plot of {col}")
    plt.show()
```

### Problem 3:
#### Draw the scatter plot matrix for Male between all the variables from the following data. Consider line, histogram and density plots diagonally in separate plots.

```python
df = pd.read_excel("male.xlsx")
```

```python
import seaborn as sns
```

```python
# Create a pairplot with different kinds of plots on the diagonal
g = sns.pairplot(df, diag_kind='kde', kind='scatter', diag_kws={'fill': True, 'linewidth': 2})

# Customize the plots
for i, j in zip(*np.tril_indices_from(g.axes, -1)):
    g.axes[i, j].set_xlabel('')
    g.axes[i, j].set_ylabel('')

plt.show()
```

```python
# Second scatter plot matrix with histogram in diagonal
g2 = sns.pairplot(df, diag_kind='hist', kind='scatter')
plt.show()
```

```python
# Third scatter plot matrix with line plot in diagonal (requires a custom function for diagonal)
def diag_line(x, **kwargs):
    plt.plot(x, x, color='r', lw=2)  # Identity line for simplicity

g3 = sns.PairGrid(df)
g3.map_diag(diag_line)
g3.map_offdiag(sns.scatterplot)
plt.show()
```
